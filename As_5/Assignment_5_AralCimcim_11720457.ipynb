The cumulant-generating function $K_X(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
\newcommand\E{\operatorname{\mathbb{E}}}
\newcommand\dif{\mathrm{d}}
K_X(t) = \log E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n(X) = K_X^{(n)}(0)$ the $n$-th derivative of $K_X(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
K_X(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1 (20 Points)</h2>
Show that if the random variables $X$ and $Y$ are independent then 
$\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds. 

For the first part:

$\log (M_{X+Y}(t)) = \log (M_X(t) M_Y(t))$

$= \log (M_X(t)) + \log (M_Y(t))$

$= K_X(t) + K_Y(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!} + \sum_{n=1}^\infty \kappa_n(Y) \frac{t^n}{n!}$

$= \sum_{n=1}^\infty (\kappa_n(X) + \kappa_n(Y)) \frac{t^n}{n!} = K_{x+y}(t)$

For the second part:

$\log (M_{\alpha x}(t)) = \log (M_x(\alpha t))$ 

$= K_x(\alpha t) = \sum_{n=1}^\infty \kappa_n(X) \frac{(\alpha t)^n}{n!}$

$= \sum_{n=1}^\infty \alpha^n \kappa_n(X) \frac{t^n}{n!} = K_{\alpha x}(t)$

n-th derivative at $t=0$:

$\kappa_{n}({\alpha X}) = K_{\alpha X}^{(n)}(0) = \alpha^n K_X^{(n)}(0) = \alpha^n \kappa_n(X)$

<h2 style="color:rgb(0,120,170)">Task 4 (20 Points)</h2>

Compare PCA and ICA at the example of the `cifar-10.npy` dataset. Load the dataset, perform PCA and ICA using the implementations of the `sklearn.decomposition` package with 100 components each. Then visualize the components extracted by the two methods using the provided functions `show_image` and `scale`. What differences do you observe and how do they relate to the properties of PCA and ICA? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA

def scale(s, a=0, b=1):
    return (b - a) * (s - s.min()) / (s.max() - s.min()) + a

def show_image(img):
    plt.imshow(img.reshape(3, 32, 32).transpose((1, 2, 0)))
```

```python
data = np.load('cifar-10.npy')
```

```python
pca = PCA(n_components=100)
data_pca = pca.fit_transform(data)
data_pca_inv = pca.inverse_transform(data_pca)

ica = FastICA(n_components=100)
data_ica = ica.fit_transform(data)
data_ica_inv = ica.inverse_transform(data_ica)
```

```python
plt.figure(figsize=(20, 20))
for i in range(10):
    plt.subplot(10, 10, i + 1)
    show_image(scale(pca.components_[i]))
    plt.axis('off')
    plt.title('PCA component {}'.format(i+1))
    plt.subplot(10, 10, i + 11)
    show_image(scale(ica.components_[i]))
    plt.axis('off')
    plt.title('ICA component {}'.format(i+1))
    plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=np.arange(data_pca.shape[0]))
plt.title('PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(data_ica[:, 0], data_ica[:, 1], c=np.arange(data_ica.shape[0]))
plt.title('ICA')
plt.xlabel('IC1')
plt.ylabel('IC2')
plt.show()
```

PCA seems to maximize the variance along the components with an order (e.g. giving attention to features from most significant to least significant).

ICA seems to look for mutual independence between components without an order (e.g. equal significance for all features).

Consider the cocktail party problem. Read the left channel of the files `birds.wav`, `ocean.wav`, `frogs.wav` using the `scipy.io.wavfile` package. Then construct a full-rank $3 \times 3$ mixing matrix and produce 3 different mixings of those signals. You can listen to them using the `sounddevice` package. Finally, use `FastICA` from the `sklearn.decomposition` package to recover the original sources. Listen to the recovered sources and compare them to the originals to check if ICA was successful. *Hint:* Scale the signals to $[-1,1]$ before playing them. 

```python
import scipy.io.wavfile
import sounddevice as sd

def play_sound(file):
    fs, data = scipy.io.wavfile.read(file)
    sd.play(data, fs)
    sd.wait()

def display_data(X):
    C = X.shape[0]
    for i in range(C):
        plt.figure(figsize=(20, 10))
        plt.subplot(C, 1, i + 1)
        plt.plot(X[i,:])
        plt.tight_layout()
    plt.show()

def play_recovered_signal(signal, fs):
    normalized_signal = signal / np.max(np.abs(signal))
    sd.play(normalized_signal, fs)
    sd.wait()
```

```python
sound_files = ['ocean.wav', 'birds.wav', 'frogs.wav']

for i in sound_files:
    play_sound(i)
    sd.wait()
```

```python
signals = []
for file in sound_files:
    fs, data = scipy.io.wavfile.read(file)
    signals.append(data[:, 0])

signals = np.stack(signals, axis=-1)

A = np.random.rand(3, 3)

print(f'Check if full rank, det = {np.linalg.det(A)}')

mixed_signals = np.dot(signals, A)
```

```python
ica = FastICA(n_components=3)
recovered_signals = ica.fit_transform(mixed_signals)
recovered_signals /= np.max(np.abs(recovered_signals), axis=0)
```

```python
display_data(mixed_signals.T)
```

```python
display_data(np.array(recovered_signals).T)
```

```python
for i in range(recovered_signals.shape[1]):
    play_recovered_signal(recovered_signals[:, i], fs)
    sd.wait()
```
